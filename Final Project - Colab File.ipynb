## **Stage 2 – Data Cleaning and Pre-processing**

Handle missing values (impute or drop)

Handle duplicates


````python
# Missing values check
print("\nMissing Values (Before):")
df.isnull().sum()
````

### Data Cleaning Steps:
1. Convert `participationbegindate` to datetime.
2. Drop `is_contracted_for_cba` (constant value).
3. Drop `practiceaddress2` and `providertypelist` (high percentage of missing values).
4. Impute missing values in `specialitieslist` and `supplieslist` with 'Unknown'.

````python
# Convert 'participationbegindate' to datetime objects
df['participationbegindate'] = pd.to_datetime(df['participationbegindate'])

# Drop 'is_contracted_for_cba' as it contains only one unique value (False)
df = df.drop(columns=['is_contracted_for_cba'])

# Drop columns with a high percentage of missing values
df = df.drop(columns=['practiceaddress2', 'providertypelist'])

# Impute missing values in 'specialitieslist' and 'supplieslist' with 'Unknown'
df['specialitieslist'].fillna('Unknown', inplace=True)
df['supplieslist'].fillna('Unknown', inplace=True)

print("Data cleaning complete. Updated DataFrame info:")
df.info()
````

##Now, let's recheck the missing values to confirm the cleaning steps were successful.

````python
# Missing values check after cleaning
print("\nMissing Values (After Cleaning):")
df.isnull().sum()
````

##Next, let's investigate the `telephonenumber` column due to its high standard deviation. We'll examine its unique values and their counts to understand its data quality and consistency.

````python
# Investigate 'telephonenumber' column
print("\nUnique values and their counts in 'telephonenumber' column (top 20):")
display(df['telephonenumber'].value_counts().head(20))

# Convert 'telephonenumber' to string to handle potential formatting issues
df['telephonenumber'] = df['telephonenumber'].astype(str)

print("\nData type of 'telephonenumber' after conversion:")
print(df['telephonenumber'].dtype)
````

##To further clean the `telephonenumber` column and ensure consistency, I will standardize its format by removing any non-numeric characters. This will make it suitable for further analysis or validation if needed.

````python
# Standardize 'telephonenumber' format by removing non-numeric characters
df['telephonenumber'] = df['telephonenumber'].str.replace(r'\D', '', regex=True)

# Display the first few cleaned telephone numbers and their length
print("\nCleaned 'telephonenumber' values (first 10):\n", df['telephonenumber'].head(10))
print("\nValue counts after cleaning (top 20):\n")
display(df['telephonenumber'].value_counts().head(20))
````

## **Stage 3 – EDA and Visualizations**


## **Univariate Analysis**

## Distribution of `acceptsassignement`
This visualization shows the proportion of medical equipment suppliers that accept assignment, which is crucial for understanding their participation in Medicare.

````python
plt.figure(figsize=(7, 5))
sns.countplot(x='acceptsassignement', data=df, palette='viridis')
plt.title('Distribution of Suppliers by Acceptance of Assignment')
plt.xlabel('Accepts Assignment')
plt.ylabel('Number of Suppliers')
plt.show()
````

## Distribution of Top `practicestate`
This chart highlights the states with the highest number of medical equipment suppliers, providing geographical insights into supplier density.

````python
plt.figure(figsize=(12, 6))
sns.countplot(y='practicestate', data=df, order=df['practicestate'].value_counts().index[:10], palette='magma')
plt.title('Top 10 Practice States by Number of Suppliers')
plt.xlabel('Number of Suppliers')
plt.ylabel('Practice State')
plt.show()
````

## Distribution of Top `specialitieslist`
This visualization shows the most common specialities offered by suppliers. The 'Unknown' category, which we imputed earlier, is also visible here.

````python
plt.figure(figsize=(12, 7))
sns.countplot(y='specialitieslist', data=df, order=df['specialitieslist'].value_counts().index[:10], palette='cubehelix', hue='specialitieslist', legend=False)
plt.title('Top 10 Specialities Offered by Suppliers')
plt.xlabel('Number of Suppliers')
plt.ylabel('Speciality List')
plt.show()
````

## Distribution of Top `supplieslist`
This chart presents the most frequently supplied items, giving an idea of the common medical equipment and supplies in the dataset. The 'Unknown' category is included.

````python
# Function to truncate long labels for better visualization
def truncate_label(label, max_length=50):
    if len(label) > max_length:
        return label[:max_length-3] + '...'
    return label

# Create a temporary column with truncated labels for plotting
df_plot = df.copy()
df_plot['supplieslist_short'] = df_plot['supplieslist'].apply(truncate_label)

plt.figure(figsize=(15, 8)) # Increased figure size again for more space
sns.countplot(y='supplieslist_short', data=df_plot,
              order=df_plot['supplieslist'].value_counts().index[:10].map(truncate_label),
              palette='viridis',
              hue='supplieslist_short', # Address FutureWarning
              legend=False) # Hide legend as hue is used for color
plt.title('Top 10 Supplies Offered by Suppliers')
plt.xlabel('Number of Suppliers')
plt.ylabel('Supplies List (Truncated for Readability)')
plt.tick_params(axis='y', labelsize=10) # Adjusted label size for readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
````

## **Bivariate Analysis**
Now, let's explore the relationships between different pairs of variables to uncover more insights.

### Relationship between `practicestate` and `acceptsassignement`
This analysis will show if there's a geographical pattern in whether suppliers accept assignment, which could indicate regional differences in Medicare participation or policy.

````python
plt.figure(figsize=(14, 7))
sns.countplot(x='practicestate', hue='acceptsassignement', data=df,
              order=df['practicestate'].value_counts().index[:10], palette='pastel')
plt.title('Acceptance of Assignment by Top 10 Practice States')
plt.xlabel('Practice State')
plt.ylabel('Number of Suppliers')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Accepts Assignment')
plt.tight_layout()
plt.show()
````

### Relationship between `specialitieslist` and `acceptsassignement`
This plot will help us understand if certain specialities are more likely to accept assignment than others, which can be an indicator of their business model or patient base.

````python
plt.figure(figsize=(16, 8))
sns.countplot(y='specialitieslist', hue='acceptsassignement', data=df,
              order=df['specialitieslist'].value_counts().index[:10], palette='muted')
plt.title('Acceptance of Assignment by Top 10 Specialities')
plt.xlabel('Number of Suppliers')
plt.ylabel('Speciality List')
plt.legend(title='Accepts Assignment')
plt.tight_layout()
plt.show()
````

## Geographical Distribution of Suppliers
Let's visualize the geographical distribution of medical equipment suppliers using latitude and longitude. This can help identify clusters or regions with high supplier density.

````python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='longitude', y='latitude', data=df, hue='acceptsassignement', palette='coolwarm', alpha=0.6)
plt.title('Geographical Distribution of Medical Equipment Suppliers by Assignment Acceptance')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
````

### Average `telephonenumber` length by `practicestate`
Though not a direct 'relationship' in the traditional sense, examining the average length of cleaned `telephonenumber` by state might reveal state-specific formatting practices or data entry quirks, which can impact data quality if not standardized.

````python
# Calculate the length of the cleaned telephone numbers
df['telephonenumber_length'] = df['telephonenumber'].astype(str).apply(len)

# Group by practicestate and calculate the mean telephone number length
state_avg_phone_len = df.groupby('practicestate')['telephonenumber_length'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=state_avg_phone_len.index, y=state_avg_phone_len.values, palette='plasma')
plt.title('Average Telephone Number Length by Top 10 Practice States')
plt.xlabel('Practice State')
plt.ylabel('Average Phone Number Length')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

## **Multivariate Analysis**
Let's delve deeper into the data by examining the relationships between multiple variables simultaneously.

## Geographical Distribution of Top Specialties
This visualization will show how different specialties are distributed across geographical regions (latitude and longitude), helping to identify clusters of specific services. I will pick the top 5 specialties to keep the plot readable.

````python
# Get the top 5 specialities
top_5_specialities = df['specialitieslist'].value_counts().index[:5]

# Filter the DataFrame to include only these top specialties
df_top_specialities = df[df['specialitieslist'].isin(top_5_specialities)]

plt.figure(figsize=(15, 10))
sns.scatterplot(x='longitude', y='latitude', hue='specialitieslist',
                data=df_top_specialities, palette='tab10', alpha=0.7, s=50)
plt.title('Geographical Distribution of Top 5 Medical Equipment Specialties')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

## Impact of `practicestate` and `specialitieslist` on `telephonenumber_length`
Let's examine if there's any notable variation in the average length of telephone numbers when grouped by both state and specialty. This can further highlight data entry or formatting conventions.

````python
# Group by top 5 practicestate and top 5 specialitieslist and calculate the mean telephone number length
# First, filter for top states and specialities to keep the plot manageable
top_states = df['practicestate'].value_counts().index[:5]
top_specialities = df['specialitieslist'].value_counts().index[:5]

df_filtered = df[df['practicestate'].isin(top_states) & df['specialitieslist'].isin(top_specialities)]

# Calculate the mean telephone number length for these groups
multi_group_avg_phone_len = df_filtered.groupby(['practicestate', 'specialitieslist'])['telephonenumber_length'].mean().unstack()

plt.figure(figsize=(16, 9))
sns.heatmap(multi_group_avg_phone_len, annot=True, fmt=".1f", cmap="YlGnBu", linewidths=.5)
plt.title('Average Telephone Number Length by Practice State and Top Specialties')
plt.xlabel('Speciality List')
plt.ylabel('Practice State')
plt.tight_layout()
plt.show()
````
